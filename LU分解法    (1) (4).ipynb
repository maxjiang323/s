## 1. LU分解法

### (1)教授寫的程式碼(有的註解是我自己加的，方便我記錄及理解)

[曾正男教授的LU分解教學影片連結](https://www.youtube.com/watch?v=W1elv5pS9xg)

```python
import numpy as np
```

```python
def LU_1(A):
    m, n = A.shape # array.shape 可以看array的形狀 ; m-->rows ; n-->columns
    U = A.copy()
    L = np.mat(np.identity(m)) # L最初的型態為單位矩陣
    P = L.copy()
    
    for i in range(m):
        
        maxEc = abs(U[i, i])
        maxRow = i
        for k in range(i+1, m):
            if abs(U[k, i]) > maxEc:
                maxEc = U[k, i]
                maxRow = k
        # swap maximum row with current row
        # [rows, columns] ; 第 i 列與第 maxRow 列互換 , 從該列的第 i 行開始(省略前面從第 0 行到第 (i-1)行的互換 ∵ 都是 0, 不用互換)
        U[[i, maxRow], i:] = U[[maxRow, i], i:]
        # [rows, columns] ; L矩陣要整行都互換, 而 P矩陣也要整列都互換
        L[[i, maxRow], :] = L[[maxRow, i], :]
        P[:, [i, maxRow]] = P[:, [maxRow, i]]
        
        # gauss elimination 
        for t in range(i+1, m):
            c = -float(U[t, i])/U[i, i] # U[i, i]--> U矩陣對角線上的數值
            # delete row t by i
            U[t, i:] = U[t, i:] + c*U[i, i:] # 列運算
            # delete column i by t
            L[t:, i] = L[t:, i] - c*L[t:, t] # 行運算
    return L, U, P
```

```python
def LU_2(A):
    m, n = A.shape
    U = np.array(A.copy())
    L = np.mat(np.identity(m))
    P = L.copy()
    
    for i in range(m):
         for t in range(i, m-1): # from the first row to (the second to last), from the second row to (the second to last)...... 
            for h in range(t+1, m): # from the second row to the last, from the third row to the last......
                U = np.mat(U)
                if U[t, i] == 0:
                    continue
                else:
                    c = -(U[h ,i] / U[t, i])
                    U[h, :] = U[h, :] + c * U[t,:]
                    L[:, t] = L[:, t] - c * L[:, h]
                
                   
    
   
         for k in range(i+1, m):
            changeRow = i
            U = np.array(U)
            nonzero = np.count_nonzero(U, axis = 1) # row
            if nonzero[k] > nonzero[i]:
                changeRow = k
                U = np.mat(U)
                U[[changeRow, i], :] = U[[i, changeRow], :]
                P[[changeRow, i], :] = P[[i, changeRow], :]
                L[[changeRow, i], :] = L[[i, changeRow], :]
                L[:, [changeRow, i]] = L[:, [i, changeRow]]
                
                
    
       
    U = np.mat(U)
    return L, U, P
       
```

```python
def LU_3(A):
    m, n = A.shape
    U = A.copy()
    U = U.astype('float64')
    P = np.mat(np.identity(m))
    L = P.copy()
    
    for i in range(m-1):
        for x in range(i+1, m):
            if U[i, i] == 0:
                if U[x, i] != 0:
                    U[[x, i], i:] = U[[i, x], i:]
                    P[[x, i], :] = P[[i, x], :]
                    L[[x, i], :] = L[[i, x], :]
                    L[:, [x, i]] = L[:, [i, x]]
        
        for t in range(i+1, m):
            if U[i, i] == 0:
                continue
            c = -(U[t, i]/U[i, i])
            U[t, i:] = U[t, i:] + c*U[i, i:]
            L[t, i] = L[t, i] - c*L[t, t]
            
    return L, U, P
```

### 說明:此前我在手寫筆記的心得中寫到$LU$分解後可以用來計算高階矩陣的行列式，但我沒有考慮到那是在$P$矩陣是單位矩陣的情況下才必定成立的結果(就是 $det(A) = \prod_{i = 0}^nU_{(i, i)}, where\;A\;is$ $a\;nxn\;matrix$)。在$P$矩陣有進行列對調的情況下，有可能出現$det(P) = -1$的情況，使式子變成$det(A) = -\prod_{i = 0}^nU_{(i, i)}, where\;A\;is$ $a\;nxn\;matrix$。所以$A$的行列式應該以
$$ A\;is\;a\;nxn\;matrix,\;det(A)= \begin{cases} \prod_{i = 0}^nU_{(i, i)}\;, & \text {if$\;det(P)=1$} \\ -\prod_{i = 0}^nU_{(i, i)}, & \text{if$\;det(P)=-1$} \end{cases} $$

### (1)利用$LU$分解法:接下來我以一個 $4$ x $4$ 的矩陣 $A\;$($A$的$P$矩陣的行列式值為$1$)$\;$及一個 $5$ x $5$ 的矩陣 $B\;$($B$的$P$矩陣的行列式值為$-1$)$\;$為例子，利用我寫的程式碼來計算它們的行列式值。

```python
import numpy as np
import sympy as sp
```

```python
def det_LU(A):
    # LU Decomposition
    m, n = A.shape
    U = A.copy()
    U = U.astype('float64')
    P = np.mat(np.identity(m))
    L = P.copy()
    count = 0
    
    for i in range(m-1):
        for x in range(i+1, m):
            if U[i, i] == 0:
                if U[x, i] != 0:
                    U[[x, i], i:] = U[[i, x], i:]
                    P[[x, i], :] = P[[i, x], :]
                    L[[x, i], :] = L[[i, x], :]
                    L[:, [x, i]] = L[:, [i, x]]
                    count = count + 1
        
        for t in range(i+1, m):
            if U[i, i] == 0:
                continue
            c = -(U[t, i]/U[i, i])
            U[t, i:] = U[t, i:] + c*U[i, i:]
            L[t, i] = L[t, i] - c*L[t, t] 

                
    det_U = 1
    for j in range(m):
        det_U = det_U * U[j, j]
        
    # check count is an even or odd to ascertain its determinant(1 or -1)
    if count % 2 == 1:
        det_P = -1
    else:
        det_P = 1
    
    # check whether det_U is going to multiply (-1) or not
    if det_P == -1:
        det_U = (-1)*det_U
    
    return det_P, det_U, L, U, P
```

```python
A = np.mat([[1, 3, 4, 5], [0, -11, 33, 7], [81, 15, -15, 2], [34, 1, 10, 2]])
print('A矩陣:')
A
```

```python
det_P, det_U, L, U, P = det_LU(A)
```

```python
print('A矩陣的P矩陣的行列式值:')
det_P
```

```python
print('A矩陣的行列式值為')
det_U
```

```python
import numpy as np
import sympy as sp
```

```python
def det_LU_sympy(A):
    # LU Decomposition
    m, n = A.shape
    U = A.copy()
    P = sp.eye(m)
    L = P.copy()
    count = 0
    
    for i in range(m-1):
        for x in range(i+1, m):
            if U[i, i] == 0:
                if U[x, i] != 0:
                    L = np.mat(L)
                    U = np.mat(U)
                    P = np.mat(P)
                    U[[x, i], i:] = U[[i, x], i:]
                    P[[x, i], :] = P[[i, x], :]
                    L[[x, i], :] = L[[i, x], :] # rpw interchange
                    L[:, [x, i]] = L[:, [i, x]] # column interchange
                    count = count + 1
                    L = sp.Matrix(L)
                    U = sp.Matrix(U)
                    P = sp.Matrix(P)
                    # for g in range(m):
                        # for h in range(m):
                            # L[g, h] = sp.Rational(L[g, h]).limit_denominator()
                            # U[g, h] = sp.Rational(U[g, h]).limit_denominator()
                    

        
        for t in range(i+1, m):
            if U[i, i] == 0:
                continue
            c = -(U[t, i]/U[i, i])
            U[t, i:] = U[t, i:] + c*U[i, i:]
            L[t, i] = L[t, i] - c*L[t, t]    
    det_U = 1
    for j in range(m):
        det_U = det_U * U[j, j]
        
    # check count is an even or odd to ascertain its determinant(1 or -1)
    if count % 2 == 1:
        det_P = -1
    else:
        det_P = 1
    
    # check whether det_U is going to multiply (-1) or not
    if det_P == -1:
        det_U = (-1)*det_U
    
    return det_P, det_U, L, U, P
```

```python
def Gauss_Elimination_1(A, b):
    m, n = A.shape # array.shape 可以看array的形狀 ; m-->rows ; n-->columns
    R = np.mat(np.zeros([m, n+1])) 
    # R is the reduced row echelon form-->簡化後的列階梯形矩陣 ; n+1-->要在A矩陣中放入b, 即改成增廣矩陣
    R[:,:n] = A # [rows, columns] ; [:, :n]-->rows:all ; cloloumns:0 ~ (n-1)
    R[:,n] = b # [rows, columns] ; [:, n]-->rows:all ; cloloumn:n
    
    for i in range(m):
        # find maximal element in the column i
        maxEc = abs(R[i, i]) # maxEc為 A矩陣對角線上的值
        maxRow = i
        for k in range(i+1, m):
            if abs(R[k, i]) > maxEc:
                maxEc = abs(R[k, i])
                maxRow = k
                
        # swap maximum row with current row
        R[[i, maxRow], i:] = R[[maxRow, i], i:] 
        # [rows, columns] ; 第 i 列與第 maxRow 列互換 , 從該列的第 i 行開始(省略前面從第 0 行到第 (i-1)行的互換 ∵ 都是 0, 不用互換) 
        
        # make all rows below this one to 0
        for k in range(i+1, m):
            c = -float(R[k, i])/R[i, i] # float要加, 避免發生計算錯誤
            R[k, i:] = R[k, i:] + c*R[i, i:]
    # solve equation for an upper triangular matrix(the numbers in the lower triangular matrix are all solved to 0)
    x = np.mat(np.zeros([m, 1]))
    for i in range(m-1, -1, -1): # range(m-1, -1, -1)-->(m-1) ~ 0
        x[i] = float(R[i, -1])/R[i, i] # 最後一行(b) / 對角線上對應的變數係數 [一元一次方程式]
        # x[i]即是[i][0](只有一行不一定要寫) ; np.matrix形式可以用index去改變數值 ; R[i][i]在矩陣的對角線上,不包含 b 的部分, 只有矩陣 A
        # 事實上 , 上三角不會被換成 0 最終目標是解出未知數 , 所以從最右列的最下面那行的一元一次方程式開始往上代入 , 以求出上面的其他未知數
        for k in range(i-1, -1, -1):  # range(i-1, -1, -1)-->(i-1) ~ 0
            R[k, -1] -= R[k, i]*x[i] 
                     
    return x 
```

### <2> 我修改後的的程式碼

```python
import numpy as np
import sympy as sp
```

```python
def Gauss_Elimination_2(A, b):
    m, n = A.shape
    # LU decomposition
    delta = A.copy()
    
    delta_x = sp.zeros(m, n)
    delta_x[:, :n] = A
    delta_x[:, 0] = b

    delta_y = sp.zeros(m, n)
    delta_y[:, :n] = A
    delta_y[:, 1] = b
    
    if m == 3:
        delta_z = sp.zeros(m, n)
        delta_z[:, :n] = A
        delta_z[:, 2] = b
    
    P1 = sp.eye(m)
    P2 = P1.copy()
    P3 = P2.copy()
    P4 = P3.copy()
    count_delta = 0
    count_delta_x = 0
    count_delta_y = 0
    count_delta_z = 0
    
    for i in range(m-1):
        for x in range(i+1, m):
            # delta
            if delta[i, i] == 0:
                if delta[x, i] != 0:
                    delta = np.mat(delta)
                    P1 = np.mat(P1)
                    delta[[x, i], i:] = delta[[i, x], i:]
                    P1[[x, i], :] = P1[[i, x], :]
                    count_delta = count_delta + 1
                    delta = sp.Matrix(delta)
                    P1 = sp.Matrix(P1)
                    
            # delta x 
            if delta_x[i, i] == 0:
                if delta_x[x, i] != 0:
                    delta_x = np.mat(delta_x)
                    P2 = np.mat(P2)
                    delta_x[[x, i], i:] = delta_x[[i, x], i:]
                    P2[[x, i], :] = P2[[i, x], :]
                    count_delta_x = count_delta_x + 1
                    delta_x = sp.Matrix(delta_x)
                    P2 = sp.Matrix(P2)
                    
            # delta y
            if delta_y[i, i] == 0:
                if delta_y[x, i] != 0:
                    delta_y = np.mat(delta_y)
                    P3 = np.mat(P3)
                    delta_y[[x, i], i:] = delta_y[[i, x], i:]
                    P3[[x, i], :] = P3[[i, x], :]
                    count_delta_y = count_delta_y + 1
                    delta_y = sp.Matrix(delta_y)
                    P3 = sp.Matrix(P3)
                    
            # delta z
            if m == 3:
                if delta_z[i, i] == 0:
                    if delta_z[x, i] != 0:
                        delta_z = np.mat(delta_z)
                        P4 = np.mat(P4)
                        delta_z[[x, i], i:] = delta_z[[i, x], i:]
                        P4[[x, i], :] = P4[[i, x], :]
                        count_delta_z = count_delta_z + 1
                        delta_z = sp.Matrix(delta_z)
                        P4 = sp.Matrix(P4)
            
                    
        
        for t in range(i+1, m):
            # delta
            if delta[i, i] == 0:
                continue
            c = -(delta[t, i]/delta[i, i])
            delta[t, i:] = delta[t, i:] + c*delta[i, i:]
            
            # delta x
            if delta_x[i, i] == 0:
                continue
            c = -(delta_x[t, i]/delta_x[i, i])
            delta_x[t, i:] = delta_x[t, i:] + c*delta_x[i, i:]
            
            # delta y
            if delta_y[i, i] == 0:
                continue
            c = -(delta_y[t, i]/delta_y[i, i])
            delta_y[t, i:] = delta_y[t, i:] + c*delta_y[i, i:]
            
            # delta z
            if m == 3:
                if delta_z[i, i] == 0:
                    continue
                c = -(delta_z[t, i]/delta_z[i, i])
                delta_z[t, i:] = delta_z[t, i:] + c*delta_z[i, i:]


    # calculate Δ, Δx, Δy, Δz
    det_1 = 1 # Δ
    det_2 = 1 # Δx
    det_3 = 1 # Δy
    det_4 = 1 # Δz
    for j in range(m):
        det_1 = det_1 * delta[j, j]
        det_2 = det_2 * delta_x[j, j]
        det_3 = det_3 * delta_y[j, j]
        if m == 3:
            det_4 = det_4 * delta_z[j, j]
        
    # check count is an even or odd to ascertain its determinant(1 or -1)
    # Δ
    if count_delta % 2 == 1:
        det_P1 = -1
        
    else:
        det_P1 = 1
        
    if det_P1 == -1:
        det_1 = (-1)*det_1

    # Δx
    if count_delta_x % 2 == 1:
        det_P2 = -1
        
    else:
        det_P2 = 1
        
    if det_P2 == -1:
        det_2 = (-1)*det_2
    
    # Δy
    if count_delta_y % 2 == 1:
        det_P3 = -1
        
    else:
        det_P3 = 1
        
    if det_P3 == -1:
        det_3 = (-1)*det_3
    
    # Δz
    if m == 3:
        if count_delta_z % 2 == 1:
            det_P4 = -1
        
        else:
            det_P4 = 1
        
        if det_P4 == -1:
            det_4 = (-1)*det_4
    
    
    
    # Gauss Elimination(all contents below this line)
    R = sp.Matrix(np.zeros([m, n+1]))
    R[:, :n] = A
    R[:,  n] = b
    
    # a unique solution(一組解)
    # delta ≠ 0
    if det_1 != 0: 
        for i in range(m):
            for v in range(i+1, m):
                if R[i, i] == 0:
                    if R[v, i] != 0:
                        R = np.mat(R)
                        R[[v, i], :] = R[[i, v], :]
                        R = sp.Matrix(R)
                        
            for h in range(i+1, m):
                if R[i, i] == 0:
                    continue
                c = R[h, i]/R[i, i]
                R[h, :] = R[h, :] - c*R[i, :]
        x = sp.zeros(m, 1) 
        for z in range(m-1, -1, -1): # (m-1) ~ 0; index = (-1) --> the last one
            x[z] = R[z, -1] / R[z, z]
            for y in range(z-1, -1, -1):
                R[y, -1] -= R[y, z] * x[z] 
            # 倒車往上消去未知數(先把A矩陣在主對角線下的元素全部消成0, 然後解好最下面的一元方程式, 往上代入二元方程式, 三元方程式, ...)
                state = '恰有一解'
    
    # no solution(無解)
    # delta = 0, one (or more) of delta_x and delta_y is(are) ≠ 0  [two-dimensional]
    # delta = 0, one (or more) of delta_x, delta_y and delta_z is(are) ≠ 0  [three-dimensional]
    if m == 2 and det_1 == 0: # two-dimensional
        if det_2 != 0 or det_3 != 0:
            state = '無解'
            x = ''
    
    if m == 3 and det_1 == 0:
        if det_2 != 0 or det_3 != 0 or det_4 != 0:
            state = '無解'
            x = ''
    
    # no solution or infinitely many solutions(無解, 或是無限多組解)
    # delta = delta_x = delta_y = 0 → infinitely many solutions [two-dimensional]
    # delta = delta_x = delta_y = delta_z = 0 → no solution or infinitely many solutions [three-dimensional]
    if m == 2 and det_1 == 0: # two-dimensional
        state = '無限多組解'
        x = ''
        
    if m == 3 and det_1 == 0: # three-dimensional
        K = sp.Matrix(np.zeros([m, n+1]))
        K[:, :n] = A
        K[:,  n] = b
        # first part for Gauss Elimination
        for i in range(m):
            for v in range(i+1, m):
                if K[i, i] == 0:
                    if K[v, i] != 0:
                        K[[v, i], :] = K[[i, v], :]
                
            for h in range(i+1, m):
                
                if K[i, i] == 0:
                    continue
                c = K[h, i]/K[i, i]
                K[h, i:] = K[h, i:] - c*K[i, i:]    
        
        
        # second part for Gauss Elimination
        for f in range(m-1, -1, -1):
            if K[f, f] == 0:
                continue
            c = K[f-1, f] / K[f, f]
            K[f-1, :] = K[f-1, :] - c*K[f, :]
        # 倒車往上，在幾個情況下對角線上元素皆為0, 只有往下的高斯消去法無法將矩陣簡化成可以判斷解情況的形式
        # example: 
        #          [[1 2 3 1] → first function
        #           [1 2 4 2] → second function
        #           [1 2 5 3]] → third function
        
        # after first part for Gauss Elimination:
        #         [[1 2 3 1]  → the number of 'nonzero elements' is four
        #          [0 0 1 1]  → the number of 'nonzero elements' is two
        #          [0 0 2 2]] → the number of 'nonzero elements' is two
        # However, this linear system has infinitely many solutions, but none of the number of 'nonzero elements' in each
        # row is 'zero'???
        
        # --> Need more Gauss Elimination process
        # after second part for Gauss Elimination:
        #         [[1 2 3 1]
        #          [0 0 0 0] → the number of 'nonzero elements' is zero(valid) 
        #          [0 0 2 2]] 
        
        # key point to ascertain the linear system has 'no solution' or 'infinitely many solutions'
        # use the number of zero in each row of K
        zero = np.count_nonzero(K, axis = 1) # count the number of 'nonzero elements' in each row of K
        t = [p for p in zero]
        
        for j in t:
            if j == 0:
                state = '無限多組解'
                x = ''
                # example: 
                #          [[2 3 4 0] → first function
                #           [1 1 1 1] → second function
                #           [1 1 1 1]] → third function
                
                # after first part for Gauss Elimination:
                #         [[2   3   4 0]
                #          [0 -0.5 -1 1]
                #          [0   0  0  0]] → the number of 'nonzero elements' is zero(valid) 
                break
            else:
                state = '無解'
                x = ''
                # example: 
                #          [[2 3 4 0] → first function
                #           [1 1 1 1] → second function
                #           [1 1 1 5]] → third function
                
                # after first part for Gauss Elimination :
                #         [[2   3   4 5]
                #          [0 -0.5 -1 1]
                #          [0   0  0  4]] → the number of 'nonzero elements' is one, not zero(valid)

                    
    return state, x
```

## (2)矩陣計算直接求解

### 如果說要偷懶一下，省去高斯消去法的計算，事實上只要確定$A$矩陣為可逆矩陣$(det(A)≠0)$，在$Ax=b$的兩邊同乘$A^{-1}$，然後就可得到:$x = A^{-1}b$，$x$矩陣中的元素即為解。不過，高斯消去法是$LU$分解法中很重要的一環，所以上面我在計算恰有一解的線性方程組時，並沒有使用這個方法，還是用高斯消去法進行計算，下面的例子我使用和上面恰有一解相同的線性方程組，看看我的程式碼算出的答案和使用矩陣計算求出來的解是否相同。

###   $\begin{cases}85x+69y-34z=120\\69x-85y+47z=10 \;\;\;--→\\23x-200y-21z=-23\end{cases} \; \left[\begin{matrix}85 & 69 & -34\\69 & -85 & 47\\23 & -200 & -21\end{matrix}\right]  \left[\begin{matrix}x\\y\\z\end{matrix}\right]  =  \left[\begin{matrix}120\\10\\-23\end{matrix}\right]$

```python
import numpy as np
import sympy as sp
```
